# **가설(hypothesis) 수립**

머신 러닝에서 식을 세울 때, 이 식을 가설이라고 한다.

보통 머신 러닝에서 가설을 임의로 추측해서 세워보는 식일 수도 있고,
경험적으로 알고 있는 식일 수도 있다.

그리고 맞는 가설이 아니라고 판단되면 계속 수정해나가게 되는 식이기도 한다.

*선형 회귀란 학습 데이터와 가장 잘 맞는 하나의 직선을 찾는 일이다.*


---



**y=Wx+b**
가설의 h를 따서 y대신 다음과 같이 식을 표현

**H(x)=Wx+b**
이때 x와 곱해지는 W를 가중치(Weight)라고 하며, b를 편향(bias)라고 한다.

# **옵티마이저 - 경사 하강법(Gradient Descent)**
여기서는 비용함수(cost function)의 값을 최소로 하는 W와 b를 찾는 방법에 대해 알아본다.

이떄 사용되는 것이 옵티마이저(Optimizer) 알고리즘(최적화 알고리즘)이다.

이 옵티마이저 알고리즘을 통해 적절한 W와 b를 찾아내는 과정을 머신 러닝에서 학습(training)이라고 한다.

여기서는 가장 기본적인 옵티마이저 알고리즘인 경사 하강법(Gradient Descent)에 대해 배운다.스크린샷 2022-03-05 오후 4.35.56.png

이번 설명에서 편향 b는 고려하지 않는다.

y=Wx와 같은 식을 기준으로 설명한다.

가중치 W가 직선의 방정식에서는 기울기다. 

위의 그림은 y=20x, y=x에 해당하는 직선을 보여준다.

기울기가 지나치게 크면 실제값과 예측값의 오차가 커지고, 

기울기가 지나치게 작아도 실제값과 예측값의 오차가 커진다.

이는 b또한 마찬가지. b가 지나치게 크거나 작으면 오차가 커진다.



---

결국 접선의 기울기가 0이 되는 지점을 찾는 것.

미분값이 0이 되는 지점. 

경사 하강법의 아이디어는 비용 함수(cost function)을 미분하여 현재 w에서의 접선의 기울기를 구하고, 접선의 기울기가 낮은 방향으로 W의 값을 변경하는 작업을 반복하는 것에 있다.




*   기울기가 음수일 때: W의 값이 증가
    
    W - a x (음수 기울기) = W + a x (양수 기울기)
*   기울기가 양수일 때: W의 값이 감소

    W - a x (양수 기울기)


스크린샷 2022-03-05 오후 4.50.57.png

a = 학습률(learning rate)

학습률 a는 W의 값을 변경할 때, 얼마나 크게 변경할 지를 결정한다.

또는 W를 그래프의 한 점으로 보고 접섭의 기울기가 0일 때까지 경사를 따라 내려간다는 관점에서는 얼마나 큰 폭으로 이동할 지를 결정한다.


```python
# 기본 셋팅
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim

# 현재 실습하고 있는 파이썬 코드를 재실행해도 다음에도 같은 결과가 나오도록 랜덤 시드를 준다.
torch.manual_seed(1)
```

```python
# 변수 선언
x_train=torch.FloatTensor([[1],[2],[3]])
y_train=torch.FloatTensor([[2],[4],[6]])
print(x_train) 
print(x_train.shape) # 크기=(3x1)
print(y_train)
print(y_train.shape) # 크기=(3x1)
```

```python
# 가중치와 편향의 초기화
# 가중치 W를 0으로 초기화하고 학습을 통해 값이 변경되는 변수임을 명시함.
W=torch.zeros(1,requires_grad=True)
print(W)

# 아래의 결과를 보고 가중치 w가 0으로 초기화되어있는 것을 알 수 있다.
# requires_grad=Ture: 이는 이 변수는 학습을 통해 계속 값이 변경되는 변수임을 의미.
```

```python
b=torch.zeros(1,requires_grad=True)
print(b)
# 지금 상태에서는 x에 어떤 값이 들어가도 가설은 0을 예측하게 된다. 
# 즉, 아직 적절한 W와 b는 아니다.
```

```python
# 가설 세우기
# H(x) = Wx+b
hypothesis=x_train*W+b
print(hypothesis)
```

```python
# 비용 함수 선언하기
# 앞서 배운 torch.mean으로 평균을 구한다.
cost=torch.mean((hypothesis - y_train)**2)
print(cost)
```

```python
# 경사 하강법 구현하기
# SGD는 경사 하강법의 일종. lr은 학습률을 의미.
optimizer=optim.SGD([W,b],lr=0.01)

# optimizer.zero_glad()를 실행하므로써 미분을 통해 얻은 기울기를 0으로 초기화한다.
# 기울기를 초기화해야만 새로운 가중치 현향에 대해서 새로운 기울기를 구할 수 있다.
# 그 다음 cost.backward() 함수를 호출하면 가중치 w에 편향 b에 대한 기울기가 계산된다.
# 그 다음 경사 하강법 최적화 함수 optimizer의 .step() 함수를 호출하여 인수로 들어갔던 W와 b에서
# 리턴되는 변수들의 기울기에 학습률 0.01을 곱하여 빼줌으로써 업데이트한다.

optimizer.zero_grad()
cost.backward()
optimizer.step()
```

```python
# 데이터
x_train=torch.FloatTensor([[1],[2],[3]])
y_train=torch.FloatTensor([[2],[4],[6]])

# 모델 초기화
W=torch.zeros(1, requires_grad=True)
b=torch.zeros(1, requires_grad=True)

# optimizer 설정
optimizer=optim.SGD([W,b],lr=0.01)

nb_epochs=1999 # 원하는 만큼 경사 하강법을 반복
for epoch in range(nb_epochs+1):
  # H(x) 계산
  hypothesis=x_train*W+b
  
  # cost 계산
  cost=torch.mean((hypothesis-y_train)**2)

  # cost로 H(x) 개선
  optimizer.zero_grad()
  cost.backward()
  optimizer.step()

  if epoch % 100 == 0:
    print('Epoch {:4d}/{} W: {:.3f}, b: {:3f} Cost: {:.6f}'.format(
        epoch,nb_epochs,W.item(),b.item(),cost.item()
    ))
```

epoch는 전체 훈련 데이터가 학습에 한 번 사용된 주기를 말한다.

이번 실습의 경우 2,000번을 수행했다.

최종 훈련 결과를 보면 최적의 기울기 W는 2에 가깝고, b는 0에 가까운 것을 볼 수 있다.

현재 훈련 데이터가 x_train은 [[1],[2],[3]]이고 y_train은 [[2],[4],[6]]인 것을 감안하면, 

실제 정답은 W가 2이고, b가 0인 H(x) = 2x이므로 거의 정답을 찾은 셈이다.

```python
# optimizer.zero_grad()가 필요한 이유
# 파이토치는 미분을 통해 얻은 기울기를 이전에 계산된 기울기 값에 누적시키는 특징이 있다.

import torch
w=torch.tensor(2.0, requires_grad=True)

nb_epochs=20
for epoch in range(nb_epochs+1):
  z=2*w
  z.backward()
  print('수식을 w로 미분한 값: {}'. format(w.grad))

# 아래의 결과를 보면 계속해서 미분값인 2가 누적되는 것을 볼 수 있다.
# 그렇기 때문에 optimizer.zero_grad()를 통해 미분값을 계속 0으로 초기화시켜줘야 한다.
```

```python
# torch.manual_seed()를 하는 이유
# torch.manual_seed()를 사용한 프로그램의 결과는 다른 컴퓨터에서 실행시켜도
# 동일한 결과를 얻을 수 있다.
# 그 이유는 torch.manual_seed()는 난수 발생 순서와 값을 동일하게 보장해준다는 특징 때문이다.
# 우선 랜덤 시드가 3일 때, 두 번 난수를 발생시켜보고, 다른 랜덤 시드를 사용한 후에 
# 다시 랜덤 시드를 3을 사용한다면 난수 발생값이 동일하게 나오는지 보겠다.

import torch
torch.manual_seed(3)
print('랜덤 시드가 3일 때')
for i in range(1,3):
  print(torch.rand(1))
```
